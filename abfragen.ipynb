```python
import spotipy
import spotipy.util as sputil

from spotipy import SpotifyClientCredentials
```

```python
# Setup

cid = 'blabla'
secret = 'blabla'
redirect = 'http://localhost:8899/'
client = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client)
```

# Search tracks + Metadata

```python
# Abfrage der 50 belibtesten Tracks vom Künstler Tristam

tracks = []
for i, track in enumerate(sp.search(q='artist:Tristam', type='track', limit=50)['tracks']['items']):
    tracks.append(track)
```

# User History

```python
# Tokenerstellung (kann auch über Browser erstellt werden und ans Backend gesendet werden)

username = 'zhgabo'
scope = 'user-read-recently-played user-top-read'

token = sputil.prompt_for_user_token(username,
                                     scope,
                                     client_id=cid,
                                     client_secret=secret,
                                     redirect_uri=redirect)

sp = spotipy.Spotify(auth=token)
```

```python
# Abfrage der letzten 50 Tracks

recently_played = []
for i, track in enumerate(sp.current_user_recently_played(limit=50)['items']):
    recently_played.append(track)

len(recently_played)
```
